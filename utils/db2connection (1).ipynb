```python
import ibm_db
```

```python
#Enter the values for you database connection

dsn_driver = "IBM DB2 ODBC DRIVER"
dsn_database = ""            # e.g. "BLUDB"
dsn_hostname = "" # e.g.: "awh-yp-small03.services.dal.bluemix.net"
dsn_port = ""                # e.g. "50000" 
dsn_protocol = "TCPIP"            # i.e. "TCPIP"
dsn_uid = ""        # e.g. "dash104434"
dsn_pwd = ""       # e.g. "7dBZ3jWt9xN6$o0JiX!m"
```

```python
dsn = (
    "DRIVER={{IBM DB2 ODBC DRIVER}};"
    "DATABASE={0};"
    "HOSTNAME={1};"
    "PORT={2};"
    "PROTOCOL=TCPIP;"
    "UID={3};"
    "PWD={4};"
    "Security={5};").format(dsn_database, dsn_hostname, dsn_port, dsn_uid, dsn_pwd,"SSL")

conn = ibm_db.connect(dsn, "", "")
```

```python
# test query
query = "SELECT count(*) FROM DB2INST1.SENSORDATA ;"
# run direct SQL
stmt = ibm_db.exec_immediate(conn, query)
ibm_db.fetch_both(stmt)
```

```python
import pandas
import ibm_db_dbi
pconn = ibm_db_dbi.Connection(conn)
df = pandas.read_sql('SELECT * FROM DB2INST1.SENSORDATA WHERE "Humidity" > 90 LIMIT 10', pconn)
df
```
